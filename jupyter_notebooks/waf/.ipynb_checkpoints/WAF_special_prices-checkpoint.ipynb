```python
import pandas as pd
```

```python
sales_cust = pd.read_csv("./csv_files/sales_customer.csv")
sales_cust = sales_cust.drop(columns=['add_1','add_2','contact','company','country','county','dnprices','email','email2','fax','inv_opt','memo','phone_1','phone_2','post_code','town','vat_num','website','zid'])
sales_cust = sales_cust.rename(columns = {'id':'custid'})
sales_cust
```

```python
works_prod = pd.read_csv("./csv_files/works_product.csv")
works_prod = works_prod.drop(columns=['short_description','description','print_order','per','unit','vat_rate','shelf_life_days','product_zid','category_id'])
works_prod = works_prod.rename(columns = {'product':'product_id'})
works_prod = works_prod.rename(columns = {'id':'prodid'})
works_prod
```

```python
sales_cupr = pd.read_csv("./csv_files/sales_customerproduct.csv")
sales_cupr
```

```python
spe_prices = pd.read_csv("./csv_files/special_prices.csv")
spe_prices = spe_prices.drop(columns=['percent','note','ts_edit','ts_add','user_add_ref','user_edit_ref','auto_updates','new_price'])
spe_prices = spe_prices.rename(columns = {'customer_ref':'customer_id', 'product_ref':'product_id'})
spe_prices
```

```python
data1 = pd.merge(works_prod, spe_prices,
how='inner', on='product_id')
data1 = data1.drop(columns=['price_x'])
data1 = data1.rename(columns = {'id':'prod_id', 'price_y':'price'})
data1
```

```python
data = pd.merge(sales_cust, data1,
how='inner', on='customer_id')
data = data.drop(columns=['customer_id','product_id'])
data = data.rename(columns = {'custid':'customer_id', 'prodid':'product_id', 'price':'ppu'})
data['order_count'] = 0
data['special_price'] = 'TRUE'
data['update_option'] = 'MANUAL'
data
```

```python
data.to_csv("./csv_files/data.csv")
```
